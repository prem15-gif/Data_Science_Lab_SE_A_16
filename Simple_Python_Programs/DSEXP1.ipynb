```python
num = int(input("Enter a number: "))
fact = 1

for i in range(1, num + 1):
    fact = fact * i

print("Factorial of", num, "is", fact)
```

```python
num = int(input("Enter a number: "))

print("Multiplication Table of", num)
for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("The number is Even")
else:
    print("The number is Odd")

```

```python
num = input("Enter a number: ")

if num == num[::-1]:
    print("The number is Palindrome")
else:
    print("The number is Not a Palindrome")
```

```python
print("Simple Health Expert System")

fever = input("Do you have fever? (yes/no): ")
cough = input("Do you have cough? (yes/no): ")
cold = input("Do you have cold? (yes/no): ")

if fever == "yes" and cough == "yes" and cold == "yes":
    print("Advice: You may have flu. Consult a doctor.")
elif fever == "yes" and cough == "no":
    print("Advice: Monitor your temperature and rest.")
elif cold == "yes" and cough == "yes":
    print("Advice: You may have a common cold.")
else:
    print("Advice: You seem healthy. Stay safe!")
```
